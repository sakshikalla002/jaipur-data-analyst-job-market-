```python
job_titles = ["data analyst", "business analyst", "data analytics", "reporting analyst", "mis executive"]

all_jobs = []

for title_keyword in job_titles:
    params = {
        "app_id": APP_ID,
        "app_key": APP_KEY,
        "what": title_keyword,
        "where": "jaipur",
        "results_per_page": 50
    }
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        print(f"'{title_keyword}' -> {data['count']} jobs found")
        
        for job in data["results"]:
            company_name = job.get("company", {}).get("display_name", "Not Available")
            location_name = job.get("location", {}).get("display_name", "Not Available")
            
            all_jobs.append({
                "search_keyword": title_keyword,
                "title": job.get("title", "Not Available"),
                "company": company_name,
                "location": location_name,
                "salary_min": job.get("salary_min"),
                "salary_max": job.get("salary_max"),
                "created": job.get("created"),
                "description": job.get("description", "")
            })
    else:
        print(f"Error for '{title_keyword}':", response.status_code)

print(f"\nTotal jobs collected: {len(all_jobs)}")
```

```python
# Salary columns ko numeric type mein convert karo (agar already nahi hai)
df["salary_min"] = pd.to_numeric(df["salary_min"], errors="coerce")
df["salary_max"] = pd.to_numeric(df["salary_max"], errors="coerce")

# Unrealistic salaries hata do (jaise jo 10,000 se kam hain - woh galat data hai, per-hour rate ho sakta hai)
df_clean = df[(df["salary_min"].isna()) | (df["salary_min"] >= 10000)]

print(f"Pehle rows: {len(df)}")
print(f"Cleaning ke baad rows: {len(df_clean)}")

df_clean.head()
```

```python
def categorize_experience(description):
    description = str(description).lower()
    if "fresher" in description or "0-1 year" in description or "entry level" in description or "no experience" in description:
        return "Fresher"
    elif "senior" in description or "5+ year" in description or "lead" in description:
        return "Senior"
    else:
        return "Mid-level / Not Specified"

df_clean["experience_level"] = df_clean["description"].apply(categorize_experience)

# Kitne fresher vs experienced jobs hain, dekho
df_clean["experience_level"].value_counts()
```

```python
skills_list = ["sql", "excel", "python", "tableau", "power bi", "r programming", "machine learning", "statistics"]

def extract_skills(description):
    description = str(description).lower()
    found_skills = [skill for skill in skills_list if skill in description]
    return ", ".join(found_skills) if found_skills else "Not mentioned"

df_clean["skills_found"] = df_clean["description"].apply(extract_skills)

df_clean[["title", "company", "skills_found"]].head(10)
```

```python
df_clean.to_csv("jaipur_jobs_cleaned.csv", index=False)
print("Cleaned data saved successfully!")
print(f"Total rows: {len(df_clean)}")
print(f"Columns: {list(df_clean.columns)}")
```

```python
from sqlalchemy import create_engine
import urllib

server = 'localhost'
database = 'DataAnalystProject'

params = urllib.parse.quote_plus(
    f"DRIVER={{ODBC Driver 17 for SQL Server}};"
    f"SERVER={server};"
    f"DATABASE={database};"
    f"Trusted_Connection=yes;"
)

engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

try:
    connection = engine.connect()
    print("Connection successful!")
    connection.close()
except Exception as e:
    print("Connection failed:", e)
```

```python
# df_clean ko SQL Server table mein daal do
df_clean.to_sql("jaipur_jobs", engine, if_exists="replace", index=False)

print("Data successfully loaded into SQL Server!")
print(f"Total rows loaded: {len(df_clean)}")
```

```python
from collections import Counter

# Saare skills ko ek list mein todo (comma se split karke)
all_skills = []
for skills_str in df_clean["skills_found"]:
    if skills_str != "Not mentioned":
        skills = [s.strip() for s in skills_str.split(",")]
        all_skills.extend(skills)

# Count karo har skill kitni baar aayi
skill_counts = Counter(all_skills)

# DataFrame banao taaki Excel mein export kar sakein
skills_df = pd.DataFrame(skill_counts.items(), columns=["Skill", "Count"]).sort_values("Count", ascending=False)

print(skills_df)
```

```python
# Existing Excel file mein nayi sheet add karo skills data ke saath
with pd.ExcelWriter("jaipur_jobs_for_excel.xlsx", engine="openpyxl", mode="a", if_sheet_exists="replace") as writer:
    skills_df.to_excel(writer, sheet_name="Skills Summary", index=False)

print("Skills data added to Excel!")
```
